### P1. Write a Python code to output the following graphs. (or similar ones)

````python
import matplotlib.pyplot as plt
plt.plot([0, 25, 50], [0, 80, 160])
plt.xlabel("x-axis")
plt.ylabel('y-axis')
plt.title('draw a line')
plt.show()
````

````python
plt.plot([1.0, 2.0, 3.0], [2.0, 4.0, 1.0])
plt.xlabel("x-axis")
plt.ylabel('y-axis')
plt.title('sample graph')
plt.show()
plt.show()
````

````python
x = [10, 20, 30]
y = [20, 40, 10]

x2 = [10, 20, 30]
y2 = [40, 10, 30]

plt.plot(x, y, label = "First Line")
plt.plot(x2, y2, label = "Second Line")

plt.xlabel('x - axis')
plt.ylabel('y- axis')

plt.legend()
plt.title('Our first graph')
#plt.title('Our first graph\nSubtitle1')

plt.show()
````

````python
x = [10, 20, 30]
y = [20, 40, 10]

x2 = [10, 20, 30]
y2 = [40, 10, 30]

plt.plot(x, y, linestyle = "dotted", label = "line1 doted")
plt.plot(x2, y2, linestyle = "dashed", label = "line 2 - dashed")

plt.xlabel('x - axis')
plt.ylabel('y- axis')

plt.legend()
plt.title('Our first graph')
#plt.title('Our first graph\nSubtitle1')

plt.show()
````

````python
x = [1,4,5,6,7]
y = [2,6,3,6,3 ]
plt.plot(x, y, linestyle = "dotted", color = "red")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
x = [1, 4,5,6,7]
y = [2, 6,3,6,3]
plt.scatter(x,y, color = 'b')
#show the graph
plt.show()
````
